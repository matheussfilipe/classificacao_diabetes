### Descrição:
<p style='font-size: 15px; line-height: 2; margin: 10px 50px; text-align: justify; text-indent: 35px;'> O diabetes é uma doença crônica que afeta milhões de pessoas em todo o mundo, caracterizada pela incapacidade do corpo de regular adequadamente os níveis de glicose no sangue. O diagnóstico precoce é crucial, pois permite a implementação de estratégias de tratamento e mudanças no estilo de vida que podem prevenir complicações graves, como doenças cardiovasculares e danos aos nervos. O uso de machine learning na detecção precoce do diabetes está revolucionando o campo médico, pois esses modelos podem analisar grandes volumes de dados clínicos para identificar padrões e fatores de risco que podem passar despercebidos em análises tradicionais. Com isso, é possível prever o desenvolvimento da doença em estágios iniciais e personalizar os cuidados para cada paciente.</p>

<p style='font-size: 15px; line-height: 2; margin: 10px 50px; text-align: justify; text-indent: 35px;'>Neste contexto o objetivo principal do nosso projeto é prever o diagnóstico de diabetes usando técnicas de machine learning.</p>

#### __Fonte dos dados__: https://www.kaggle.com/datasets/alexteboul/diabetes-health-indicators-dataset

## Preparando os dados

<p style='font-size: 14px; line-height: 2; margin: 10px 50px; text-align: justify; text-indent: 35px;'>Aqui todas as variáveis são númericas, mas ainda sim, precisamos utilizar o OneHotEnconder pois algumas delas são categóricas. Além disso, vou optar por escalonar as variáveis pois alguns dos modelos que iremos utilizar são influenciados pela escala dos dados. Antes disso, vamos separar a base de dados em alvo (y) e explicativas (x).</p>

```python
x = dados.drop('Diabetes_binary', axis = 1)
y = dados['Diabetes_binary']
```

#### Selecionando as melhores variáveis explicativas

<p style='font-size: 14px; line-height: 2; margin: 10px 50px; text-align: justify; text-indent: 35px;'> A análise prévia nos fornece uma ideia de quais as variáveis explicativas devemos manter no modelo. Para uma análise mais rigorosa irei extrair informações sobre quais iremos usar no modelo através do SelectKBest.</p>

```python
melhores_features = SelectKBest(score_func = chi2, k = 10)
fit = melhores_features.fit(x,y)

df_scores = pd.DataFrame(fit.scores_)
df_columns = pd.DataFrame(x.columns)

f_Scores = pd.concat([df_columns,df_scores],axis=1)
f_Scores.columns = ['Feature','Score']

f_Scores.sort_values(by = 'Score' ,ascending = False, inplace = True)

f_Scores.reset_index(drop = True, inplace = True)

f_Scores
```

<p style='font-size: 14px; line-height: 2; margin: 10px 50px; text-align: justify; text-indent: 35px;'> Vamos utilizar as 20 melhores variáveis para treinar nosso modelo. Assim, iremos retirar apenas a coluna AnyHealthcare da nossa base de dados.
</p>

```python
columns = ["AnyHealthcare"]

dados.drop(columns ,axis= 1, inplace=True)

x = dados.drop('Diabetes_binary', axis = 1)
y = dados['Diabetes_binary']
```

#### OneHotEncoder

```python
colunas = x.columns

one_hot = make_column_transformer((OneHotEncoder( drop = 'if_binary'), ['GenHlth', 'Age', 'Education', 'Income']), remainder = 'passthrough', sparse_threshold = 0)

x = one_hot.fit_transform(x)
```

#### Normalizando os dados

```python
scaler = StandardScaler()

x = scaler.fit_transform(x)
```

#### Separando em treino e teste

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, 
                                                        y, 
                                                        test_size = 0.30, 
                                                        random_state = 42, 
                                                        stratify = y)
```

<p style='font-size: 14px; line-height: 2; margin: 10px 50px; text-align: justify; text-indent: 35px;'> Para este projeto iremos escolher 6 modelos: Árvore de decisão, Random Forest, Logistic Regression, Naive Bayes, XGBoost e AdaBoost. A ideia é ver quais modelos tem os melhores desempenhos.</p>

```python
ConfusionMatrixDisplay.from_estimator(arvore, x_teste, y_teste, display_labels=['Não diabético', 'Diabético'], 
                                                                    normalize='true', cmap='Blues'); 
```

# <font color='red' style='font-size: 30px;'>Melhorando o modelo: Otimizando os hiperparâmetros</font>
<hr style='border: 2px solid red;'>

<p style='font-size: 14px; line-height: 2; margin: 10px 50px; text-align: justify; text-indent: 35px;'> Nestra etapa de melhora do modelo, vamos fazer uma otimização dos hiperparâmetros. Para isso, vamos utilizar um pipeline da biblioteca imbpipeline com dois passos: Normalização dos dados + Modelo. Além disso, vamos realizar a busca dos melhores hiperparâmetros utilizando o GridSearchCV ou RandomSearchCV fazendo uma validação cruzada com o StratifiedKFold. Nesta otimização vamos escolher como score o recall dado a importância que ele tem para o nosso problema. Entretanto, sempre vamos manter a acurácia como métrica principal de análise.

#### Árvore de decisão

```python
pipeline = imbpipeline([(('scaler', StandardScaler())), ('arvore', DecisionTreeClassifier(random_state = 42))])
param_grid_dt = {
    'arvore__criterion':  ['gini', 'entropy'],
    'arvore__max_depth': [1, 3, 6, 8],
    'arvore__min_samples_split': [2, 32, 56, 256],
    'arvore__min_samples_leaf':  [2, 32, 56, 256],
    'arvore__splitter': ['best', 'random'],
    'arvore__max_features' : [None, 'sqrt', 'log2'],
    'arvore__class_weight' : [None, 'balanced']
}

skf = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 42)

random_dt = RandomizedSearchCV(pipeline, param_grid_dt, cv = skf,
                              verbose = 1, scoring = 'recall', 
                               n_jobs = -1, random_state = 42, n_iter = 200)

random_dt.fit(x_treino, y_treino)

print(f'Recall: {random_dt.best_score_:.3f}')

print(f'Melhor modelo: {random_dt.best_params_}')
```

```python
y_previsto_dt = random_dt.predict(x_teste)

print(f' Acurácia da Árvore de decisão otimizado: {accuracy_score(y_teste, y_previsto_dt):.2f} e Recall: {recall_score(y_teste, y_previsto_dt):.2f}')
```

Como última tentativa de melhora podemos combinar esses modelos através do VotingClassfier a fim de obter um resultado ainda melhor.

```python
params_lr = {
    'solver': 'liblinear',
    'penalty': 'l1',
    'max_iter': 150,
    'C': 0.01,
    'random_state': 42
}

params_xgb = {
    'n_estimators': 25,
    'max_depth': 6,
    'learning_rate': 0.2,
    'gamma': 0.3,
    'colsample_bytree': 1
}

params_ada = {
    'max_depth': 2
}

modelo_logistica = LogisticRegression(**params_lr)
modelo_xgboost = xgb.XGBClassifier(**params_xgb)
modelo_ada = AdaBoostClassifier(estimator=DecisionTreeClassifier(**params_ada), learning_rate=1.0, n_estimators=200, random_state=42)

modelo_combinado = VotingClassifier(estimators=[('lr', modelo_logistica), ('xgb', modelo_xgboost), ('ada', modelo_ada)], voting='hard')

modelo_combinado.fit(x_treino, y_treino)

y_pred = modelo_combinado.predict(x_teste)
```

```python
print(f"Acurácia: {100*round(accuracy_score(y_teste, y_pred),4)}")
print(f"Recall: {100*round(recall_score(y_teste, y_pred),4)}")
```

<p style='font-size: 14px; line-height: 2; margin: 10px 50px; text-align: justify; text-indent: 35px;'> Após a combinação o nosso modelo obtém uma acurácia similar a do XGBoost e um recall inferior ao mesmo.
